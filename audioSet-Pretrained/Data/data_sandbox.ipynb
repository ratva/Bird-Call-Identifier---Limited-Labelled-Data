```python
import pandas as pd
import copy

df = pd.read_csv("/Users/angelashen/Halligan/l3d/train_metadata.csv")

species = ['Eurasian Coot', "Blyth's Reed Warbler", 'Gray Heron', 'Greenish Warbler', 'Black-crowned Night-Heron', 'Rose-ringed Parakeet', 'Greater Racket-tailed Drongo', 'White-breasted Waterhen', 'Gray-headed Canary-Flycatcher', 'Great Egret', 'Little Egret', 'Red-whiskered Bulbul']
mask = df['common_name'].isin(species)

species_df = df[mask]

def filter_by_unique(dataframe):
    df = copy.deepcopy(dataframe)
    df["geo_data"] = df.apply(lambda row: str(row['latitude'])+" "+str(row['longitude'])+" "+str(row['author']), axis=1)
    df = df.sort_values(by=['rating'], ascending=False)
    df = df.drop_duplicates(subset=['geo_data', 'common_name'])
    print(df.value_counts())
    return df

filtered_df = filter_by_unique(species_df)
```

```python
common_name_counts = filtered_df['common_name'].value_counts()
common_name_counts
```

```python
rating_max = filtered_df['rating'].max() 
rating_min = filtered_df['rating'].min()  
rating_range = rating_max - rating_min 

print("Maximum Rating:", rating_max)
print("Minimum Rating:", rating_min)
print("Range of Rating:", rating_range)
```
